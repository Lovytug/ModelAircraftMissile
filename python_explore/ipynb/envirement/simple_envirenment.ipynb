```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def air_density(h):
    return 1.225 * np.exp(-h / 8000)
```

```python
def air_pressure(h):
    return 101325 * np.exp(-h / 7000)
```

```python
h_max = 25 * 1e3

step_meters = 100

h = np.linspace(0, h_max, step_meters)

value_density = air_density(h)

value_pressure = air_pressure(h)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))  # figsize   

#   ( )
ax[0].plot(value_pressure, color='red', label='')
ax[0].set_xlabel(', ')
ax[0].set_ylabel(', ')
ax[0].set_title('   ')
ax[0].grid(True)
ax[0].legend()

#   ( )
ax[1].plot(value_density, color='blue', label='')
ax[1].set_xlabel(', ')
ax[1].set_ylabel(', /')
ax[1].set_title('   ')
ax[1].grid(True)
ax[1].legend()

plt.tight_layout()  #   
plt.show()
```
